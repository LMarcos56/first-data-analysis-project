####  We will see how the Production of Oil (produção de óleo) change around the years in Rio de Janeiro state

```python
df3 = full_data.loc[full_data['Estado'] == 'Rio de Janeiro']
```

```python
df3_new = df3.groupby(["Ano", "Estado"]).sum()
df3_new
```

**We will normalize the data for better visualization**

```python
normalized_df3=(df3_new-df3_new.min())/(df3_new.max()-df3_new.min())
normalized_df3
```
